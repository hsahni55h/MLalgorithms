## Algerian Forest Fires Dataset 
Data Set Information:

The dataset includes 244 instances that regroup a data of two regions of Algeria,namely the Bejaia region located in the northeast of Algeria and the Sidi Bel-abbes region located in the northwest of Algeria.

122 instances for each region.

The period from June 2012 to September 2012.
The dataset includes 11 attribues and 1 output attribue (class)
The 244 instances have been classified into fire(138 classes) and not fire (106 classes) classes.

Attribute Information:

1. Date : (DD/MM/YYYY) Day, month ('june' to 'september'), year (2012)
Weather data observations
2. Temp : temperature noon (temperature max) in Celsius degrees: 22 to 42
3. RH : Relative Humidity in %: 21 to 90
4. Ws :Wind speed in km/h: 6 to 29
5. Rain: total day in mm: 0 to 16.8
FWI Components
6. Fine Fuel Moisture Code (FFMC) index from the FWI system: 28.6 to 92.5
7. Duff Moisture Code (DMC) index from the FWI system: 1.1 to 65.9
8. Drought Code (DC) index from the FWI system: 7 to 220.4
9. Initial Spread Index (ISI) index from the FWI system: 0 to 18.5
10. Buildup Index (BUI) index from the FWI system: 1.1 to 68
11. Fire Weather Index (FWI) Index: 0 to 31.1
12. Classes: two classes, namely Fire and not Fire

## Understanding Histograms of the Dataset

### Overview

Histograms are a powerful tool for visualizing the distribution of numerical data. They display the frequency of data points within specified ranges (bins). Here, we generate histograms for each numerical feature in our dataset to understand their distributions.

### Columns in the Dataset

The dataset contains the following columns:
- **Temperature**: Temperature values.
- **RH (Relative Humidity)**: Relative humidity percentages.
- **Ws (Wind Speed)**: Wind speed values.
- **Rain**: Amount of rainfall.
- **FFMC (Fine Fuel Moisture Code)**: Indicates the moisture content of fine fuels.
- **DMC (Duff Moisture Code)**: Indicates the moisture content of decomposing organic material.
- **DC (Drought Code)**: Indicates the long-term moisture content affecting drought conditions.
- **ISI (Initial Spread Index)**: Indicates the rate of fire spread.
- **BUI (Build-Up Index)**: Indicates the availability of fuel for combustion.
- **FWI (Fire Weather Index)**: Overall indicator of fire danger.
- **Classes**: Categorical data representing fire danger classes.
- **Region**: Categorical data representing different regions.

### Key Points to Infer from Histograms

1. **Distribution Shape**:
   - **Normal Distribution**: Symmetric, bell-shaped histogram.
   - **Skewed Distribution**: Asymmetry indicates skewness (left or right).
   - **Uniform Distribution**: Similar heights for bars indicate uniform distribution.

2. **Central Tendency**:
   - **Mean**: Approximate location where most data points cluster.
   - **Median**: Middle value, particularly relevant for symmetric distributions.

3. **Spread**:
   - **Range**: Width of the histogram indicates the data range.
   - **Variance**: Wider histograms indicate higher variance.

4. **Outliers**:
   - **Outliers**: Bars separated from others indicate extreme values.

5. **Modes**:
   - **Unimodal**: Single peak.
   - **Bimodal/Multimodal**: Multiple peaks indicating subgroups.

### Example Interpretation

Here’s how to interpret the histograms for some of the features:

- **Temperature**:
  - **X-axis**: Temperature values.
  - **Y-axis**: Frequency of each temperature range.
  - **Inference**: Common temperature ranges, presence of outliers, and distribution shape.

- **Relative Humidity (RH)**:
  - **X-axis**: Humidity percentages.
  - **Y-axis**: Frequency of each humidity range.
  - **Inference**: Common humidity levels, skewness, and variance.

- **Wind Speed (Ws)**:
  - **X-axis**: Wind speed values.
  - **Y-axis**: Frequency of each wind
